```python
from ydata_profiling import ProfileReport

# profile = ProfileReport(df, 
#                         title="My Report", 
#                         explorative=True,   # 是否开启“探索模式”，更多图表
#                         minimal=True)       # 是否开启“最简模式”，更快


edu_report = ProfileReport(edu, title="EDA Report - edu.csv", minimal=True)
positions_report = ProfileReport(positions, title="EDA Report - positions.csv", minimal=True)

# 保存为 HTML 文件
edu_report.to_file("edu_profiling_report.html")
positions_report.to_file("positions_profiling_report.html")

print("✅ 报告已生成：edu_profiling_report.html, positions_profiling_report.html")
```

```python
users_report = ProfileReport(users, title="EDA Report - users.csv", minimal=True)

users_report.to_file("users_profiling_report.html")

print("✅ 报告已生成：users_profiling_report.html")
```
